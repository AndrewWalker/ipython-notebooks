```python
import networkx as nx
import numpy as np
%matplotlib inline
import functools
```

Calculate the number of paths in a DAG by depth first search / dynamic programming.

https://cs.stackexchange.com/questions/3078/algorithm-that-finds-the-number-of-simple-paths-from-s-to-t-in-g

```python
def count_paths(G, s, t):
    @functools.lru_cache()
    def dfs(u, t, depth):
        assert(depth < 7)
        if u == t:
            return 1
        else:
            return sum(dfs(c, t, depth+1) for c in G.neighbors(u))
    return dfs(s, t,0 )
```

```python
A = np.zeros((14,14))
A[0,1:4] = 1
A[1,4:7] = 1
A[2,4:7] = 1
A[3,4:7] = 1
A[4,7:10] = 1
A[5,7:10] = 1
A[6,7:10] = 1
A[7,10:13] = 1
A[8,10:13] = 1
A[9,10:13] = 1
A[10,13] = 1
A[11,13] = 1
A[12,13] = 1
A
```

```python
G = nx.from_numpy_array(A, create_using=nx.DiGraph())
nx.draw(G)
```

```python
count_paths(G, 0, 13)
```
