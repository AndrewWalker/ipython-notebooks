# Hypotrochoid

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
from sympy import Rational
from sympy.plotting import plot_parametric
```

```python
def hypotrochoid(R, r, h, t):
    xs = (R-r) * sympy.cos(t) + h * sympy.cos(((R-r)*t)/r)
    ys = (R-r) * sympy.sin(t) - h * sympy.sin(((R-r)*t)/r)
    period = ((R-r)/r).q * 2 * sympy.pi
    return xs, ys, period

def plot_hypotrochoid(R, r, h):
    t = sympy.Symbol('t')
    xs, ys, p = hypotrochoid(R, r, h, t )
    plot_parametric(xs, ys, (t, 0.0, p))
    
R = Rational(1, 5)
r = Rational(-1, 13)
h = Rational(-1, 9)
plot_hypotrochoid(R, r, h)
```
