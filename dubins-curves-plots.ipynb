# Validation and examples of the Python dubins module

```python
import dubins
import numpy
import itertools
import matplotlib.pyplot as plt
```

```python
def dubins_curve_cost_plot(theta1, domain = 6, lvl_spacing = 0.4):
    """Plot the 
    """
    q0 = (0, 0, numpy.pi/2)
    n  = 161
    xs = numpy.linspace(-domain, domain, n)
    ys = numpy.linspace(-domain, domain, n)
    L = numpy.zeros((n,n))
    for i, j in itertools.product(range(n), range(n)):
        x = xs[i]
        y = ys[j]
        L[j][i] = dubins.path_length(q0, (x, y, theta1), 1.0)
    lvls = numpy.arange(0, L.flatten().max()+lvl_spacing, lvl_spacing)
    plt.contour(xs, ys, L, levels = lvls)
    plt.colorbar()

# Equivalent to Fig 26 of Chapter 3 "Optimal Trajectories for Nonholonomic Mobile Robots" by Soueres and Boissonnat
# From the book "Robot Motion Planning and Control" edited by "Jean-Paul Laumond", which is available online at
# http://ftp.laas.fr/pub/ria/promotion/chap3.pdf
dubins_curve_cost_plot(0.0)
```
