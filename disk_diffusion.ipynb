# Diffusion in the unit circle (2d)
*Andrew Walker 2013*

Analysis of pure diffusion in a unit disk with Dirchelet boundary conditions and an initial distribution that it a dirac delta function at x=0, y=0, t=0. The result is based on an approximate expression for the short- and long- term limits of the Green's function approximation, and is based on a small part of the paper "A fast random walk algorithm for computing the pulsed-gradient spin-echo signal in multiscale porus media", by Denis S. Grebenkov

```python
import numpy
from scipy.special import *

def short_term(ts):
    return 1.0 - 2.0 * numpy.exp(-1/(4.0 * ts)) * (1.0 - ts + 4.0*ts**2)

def long_term(ts):
    alpha_00 = jn_zeros(0,1)[0]
    return 2.0/(alpha_00*jn(1,alpha_00)) * numpy.exp(-alpha_00**2 * ts)

ts = numpy.logspace(-2.0,1.0, 51)

plt.plot(ts, short_term(ts), '.', label='short term approximation')
plt.plot(ts, long_term(ts), 'r.', label='long term approximation')
plt.gca().set_xscale('log')
plt.gca().set_ylim(-0.1, 1.1)
plt.xlabel('time')
plt.ylabel('proportion surviving')
plt.title('Approximate survival function for diffusion in the unit circle')
```

```python
def truncated_survival(kmax, ts):
    # Naive implementation (no optimisations)
    ss = numpy.zeros((kmax, ts.shape[0]))
    for k in xrange(kmax):
        alpha_0k = jn_zeros(0, k+1)[-1]
        for ti, t in enumerate(ts):
            ss[k,ti] = 1 / (alpha_0k * jn(1, alpha_0k)) * numpy.exp(-alpha_0k**2 * t)
    ss = 2.0 * numpy.sum(ss, axis=0)
    return ss
    
ts = numpy.logspace(-2.0,1.0, 51)    
plt.plot(ts, truncated_survival(100, ts), 'g-', label='truncated approximation')
plt.plot(ts, short_term(ts), '.', label='short term approximation')
plt.plot(ts, long_term(ts), 'r.', label='long term approximation')

plt.gca().set_xscale('log')
plt.gca().set_ylim(-0.1, 1.1)
```
