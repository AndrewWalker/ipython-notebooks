# Diffusion in the unit circle (2d)
*Andrew Walker 2013*

Analysis of pure diffusion in a unit disk with Dirchelet boundary conditions and an initial distribution that it a dirac delta function at x=0, y=0, t=0. The result is based on an approximate expression for the short- and long- term limits of the Green's function approximation, and is based on a small part of the paper "A fast random walk algorithm for computing the pulsed-gradient spin-echo signal in multiscale porus media", by Denis S. Grebenkov

```python
import numpy
from scipy.special import *
from scipy.interpolate import piecewise_polynomial_interpolate

def short_term(ts):
    return 1.0 - 2.0 * numpy.exp(-1/(4.0 * ts)) * (1.0 - ts + 4.0*ts**2)

def long_term(ts):
    alpha_00 = jn_zeros(0,1)[0]
    return 2.0/(alpha_00*jn(1,alpha_00)) * numpy.exp(-alpha_00**2 * ts)

ts = numpy.logspace(-2.0,1.0, 51)

plt.plot(ts, short_term(ts), '.', label='short term approximation')
plt.plot(ts, long_term(ts), 'r.', label='long term approximation')
plt.gca().set_xscale('log')
plt.gca().set_ylim(-0.1, 1.1)
plt.xlabel('time')
plt.ylabel('proportion surviving')
plt.title('Approximate survival function for diffusion in the unit circle')
```

```python
def truncated_survival(kmax, ts):
    # Naive implementation (no optimisations)
    ss = numpy.zeros((kmax, ts.shape[0]))
    for k in xrange(kmax):
        alpha_0k = jn_zeros(0, k+1)[-1]
        for ti, t in enumerate(ts):
            ss[k,ti] = 1 / (alpha_0k * jn(1, alpha_0k)) * numpy.exp(-alpha_0k**2 * t)
    ss = 2.0 * numpy.sum(ss, axis=0)
    return ss
    
ts = numpy.logspace(-2.0,1.0, 51)    
plt.plot(ts, truncated_survival(100, ts), 'g-', label='truncated approximation')
plt.plot(ts, short_term(ts), '.', label='short term approximation')
plt.plot(ts, long_term(ts), 'r.', label='long term approximation')

plt.gca().set_xscale('log')
plt.gca().set_ylim(-0.1, 1.1)
```

# Numerical inversion of the Survival function

This is effectively the bisection search and linear interpolation described in the Grebhenkov paper. This is convenient for a vectorised implementation in numpy because the search works in a convenient way (searchsorted). This would be a bit more difficult to implement with a root finding algorithm (without resorting to cython).

```python
def first_passage_time(kmax = 100, tmax=51):
    # vectorised piecewise linear interpolation using
    # an approximation of `kmax` terms in the survival function
    # and a uniform grid in the time space
    ts = numpy.logspace(-2.0,1.0, tmax)
    ps = truncated_survival(kmax, ts)
    ps = ps[::-1]
    ts = ts[::-1]
   
    def impl(newps):
        # O( log[n] ) where n is the size of ps
        # which is equal to tmax
        high = numpy.searchsorted(ps, newps)
        low  = high - 1       
        # start of each linear bracket
        plow  = ps[low]
        # end if each linear bracket
        phigh = ps[high]
        # gradient of each linear segment
        ks = (newps - plow)/(phigh - plow)
        # start of each time bracket
        tlow  = ts[low]
        # end of each time bracket
        thigh = ts[high]       
        return ks *(thigh - tlow) + tlow
    return impl
 
f = first_passage_time(tmax =100)
nps = numpy.random.uniform(low = 1e-20, high = 1.0 - 1e-12, size = 10000)
nts = f(nps)
 
plt.plot(nts,nps,'.')
ts = numpy.logspace(-2.0,1.0, 51)
ps = truncated_survival(kmax=100, ts=ts)
plt.plot(ts, ps, 'g-');
plt.gca().set_xscale('log');
plt.gca().set_ylim(-0.1, 1.1);
plt.gca().set_xlim(-0.1, 11.0);
 
plt.figure();
plt.hist(nts,bins=50, normed=True);
 
```

# Monte Carlo solution of the same problem

This alternative development of the solution is based on applying knowledge of independent particles diffuse in the infinite plane. Assuming that we can overcome the limitations of this type of monte carlo simulation. This involve selecting the number of particles to be sufficiently large that you can guarantee that the error term will be small $O(1/\sqrt(n))$ for $n$ particles. The error in the first passage time estimate is $O(\Delta t)$

```python
import numpy
import numpy.random

def monte_carlo_disk(n, dt, L=1.0):
    xs = numpy.zeros((n, 2))
    D  = 1.0
    ts = []
    time = 0.0
    while xs.shape[0] > 0:
        time += dt
        sigma = numpy.sqrt(2 * D * dt)
        dx = sigma * numpy.random.normal(0.0, 1.0, size= xs.shape)
        xs += dx
        Ls = numpy.sqrt(numpy.sum(xs**2, axis = 1))
        absorbed = numpy.where(Ls >= L)[0]
        retain   = numpy.where(Ls < L)[0]
        if absorbed.shape[0] != 0:
            xs = xs[retain,:]
            ts.append([time]*absorbed.shape[0])
    return numpy.hstack(ts)

times = monte_carlo_disk(10000, 0.001)
plt.hist(times, bins = numpy.linspace(0.0, 1.0, 101), normed=True);
```
